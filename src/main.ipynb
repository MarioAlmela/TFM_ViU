```python
import sys
sys.path.append('.')
```

```python
from DataGen import DataGenerator
from GAN import GenerativeAdversarialNetwork
import tensorflow as tf
import pandas as pd
import os
import json
import matplotlib.pyplot as plt
from PIL import Image
import numpy as np
```

```python
# Hyperparameters
input_dim = (28, 28, 3)
z_dim = 100
disc_conv_filters = [64, 64]
disc_conv_kernels = [(3, 3), (3, 3)]
disc_conv_strides = [(2, 2), (2, 2)]
gen_conv_filters = [128, 128, 3]
gen_conv_kernels = [(4, 4), (4, 4), (7, 7)]
gen_conv_strides = [(2, 2), (2, 2), (1, 1)]
sess = tf.function()
lr = 0.005
batch_size = 128
epochs = 10
is_training = True
```

```python
experiment_folder = os.path.abspath(os.path.join('..', 'Experimentos'))
model_name = 'GAN'
image_path = os.path.abspath(os.path.join('..', 'Dataset'))
data_path = os.path.join(image_path, 'wclasses.csv')

model_path = os.path.abspath(os.path.join(experiment_folder, model_name))
if not os.path.exists(model_path):
    os.mkdir(model_path)

if len(os.listdir(os.path.join(experiment_folder, model_name))) == 0:
    exp_id = 1
else:
    exp_id = len(os.listdir(os.path.join(experiment_folder, model_name))) + 1

current_experiment_path = os.path.join(experiment_folder, model_name, 'Exp_'+str(exp_id))
if not os.path.exists(current_experiment_path):
    os.mkdir(current_experiment_path)
```

```python
hyperparams = {
    "input_dim":input_dim,
    "z_dim":z_dim,
    "disc_conv_filters":disc_conv_filters,
    "disc_conv_kernels":disc_conv_kernels,
    "disc_conv_strides":disc_conv_strides,
    "gen_conv_filters":gen_conv_filters,
    "gen_conv_kernels":gen_conv_kernels,
    "gen_conv_strides":gen_conv_strides,
    "learning rate":lr,
    "batch_size":batch_size,
    "epochs":epochs,
    "opt":"Adam",
    "loss":"binary_crossentropy",
    "data_path":data_path
}

with open(os.path.join(current_experiment_path, 'hyperparams.json'), 'w') as f:
    json.dump(hyperparams, f)
```

```python
# Train the model
if is_training:
    df = pd.read_csv(data_path, sep=',', encoding='ANSI')
    # Load data
    data_flow = DataGenerator(df, path_to_img=image_path, batch_size=batch_size, width=input_dim[0], height=input_dim[1], channels=input_dim[2], shuffle=True)
    gan = GenerativeAdversarialNetwork(input_dim, disc_conv_filters, disc_conv_kernels, disc_conv_strides, gen_conv_filters, gen_conv_kernels, gen_conv_strides, z_dim)
    gan.build(use_batch_norm=True, use_dropout=False, use_flatten=True)
    gan.compile(learning_rate=lr)
    gan.train(data_flow, epochs=epochs, save_path=current_experiment_path, save_interval=1, disc_train_steps=1)
```

```python
is_training = False
if not is_training:
    _, generator, gan = GenerativeAdversarialNetwork.load_model(current_experiment_path)
    # Show batch of generated images
    fig = plt.figure(figsize=(10, 10))
    rows = 2
    cols = 5
    x_input = np.random.randn(rows*cols*generator.layers[0].get_config()['batch_input_shape'][1]).reshape(rows*cols, generator.layers[0].get_config()['batch_input_shape'][1])
    imgs = generator.predict(x_input)
    for i in range(1, rows*cols+1):
        img = Image.fromarray(np.uint8(imgs[i-1])).convert('RGB')
        fig.add_subplot(rows, cols, i)
        plt.imshow(img)
    plt.show()
    print('\n')
```
